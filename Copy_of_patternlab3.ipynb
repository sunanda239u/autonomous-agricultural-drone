```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/drive/MyDrive/50_Startups.csv')
comp=['Amazon','Google','Microsoft','Apple','IBM']
df['Company']=np.random.choice(comp,size=len(df))
serialx=np.array(comp)
prft=[0,0,0,0,0]
spn=[0,0,0,0,0]
adn=[0,0,0,0,0]
rds=[0,0,0,0,0]
for i in range(5):
    c=0
    for j in range(len(df)):
        if comp[i]==df['Company'][j]:
            prft[i]=prft[i]+df['Profit'][j]
            spn[i]=spn[i]+df['Marketing Spend'][j]
            adn[i]=adn[i]+df['Administration'][j]
            rds[i]=rds[i]+df['R&D Spend'][j]
            c=c+1
    prft[i]=prft[i]/c
    spn[i]=spn[i]/c
    adn[i]=adn[i]/c
    rds[i]=rds[i]/c
print("Comapnies: ",comp)
print("Profit avg: ",prft)
print("Mrkt spend avg: ",spn)
print("Admin avg: ",adn)
print("R&D avg: ",rds)

profity=np.array(prft)
spendy=np.array(spn)
adminy=np.array(adn)
rdsy=np.array(rds)

plt.subplot(1,4,1)
plt.bar(serialx,rdsy)
plt.xlabel('R&D Spend')
plt.subplot(1,4,2)
plt.bar(serialx,adminy)
plt.xlabel('Administration')
plt.subplot(1,4,3)
plt.bar(serialx,spendy)
plt.xlabel('Marketing Spend')
plt.subplot(1,4,4)
plt.bar(serialx,profity)
plt.xlabel('Profit')
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

df=pd.read_csv('/content/drive/MyDrive/CC GENERAL.csv')

x=df.iloc[:,1:-1]
x=x.fillna(x.mean())

kval=[i for i in range(2,15)]
sscore=[]
maxs=0
bestk=0
for k in kval:
  kmeans=KMeans(n_clusters=k,random_state=13)
  kmeans.fit(x)
  y=kmeans.predict(x)
  ss=silhouette_score(x,y)
  sscore.append(ss)
  if ss > maxs:
    maxs=ss
    bestk=k
  # print("For K value: ",k," silhoutte score: ",ss)

plt.plot(kval,sscore)
plt.xlabel('K value')
plt.ylabel('Silhoutte score')
plt.show()
print("Best K value: ",bestk)
```

```python
km=KMeans(n_clusters=bestk,random_state=13)
km.fit(x)
y=km.predict(x)

plt.scatter(x['BALANCE'][y==0],x['PURCHASES'][y==0],color='blue')
plt.scatter(x['BALANCE'][y==1],x['PURCHASES'][y==1],color='green')
plt.show()
```

apply Decision tree and Random forest classifier on the given dataset and analyze accuracy using confusion matrix and fbeta score

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import fbeta_score,confusion_matrix,ConfusionMatrixDisplay,classification_report

# import matplotlib.pyplot as plt
df=pd.read_csv('/content/drive/MyDrive/patterndata.csv')
x=df.iloc[:,2:-1]
y=df.iloc[:,1]
# [rowstart:rowend , columnstart:columnend-1]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)

dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
yptrain=dt.predict(xtrain)
yptest=dt.predict(xtest)

trfb=fbeta_score(ytrain,yptrain,average='weighted',beta=1)
tsfb=fbeta_score(ytest,yptest,average='weighted',beta=1)
trcm=confusion_matrix(ytrain,yptrain)
tscm=confusion_matrix(ytest,yptest)

print("Training fbeta:",trfb)
print("Testing fbeta:",tsfb)
print("Training Confusion matrix:\n",trcm)
print("Testing confusion matrix:\n",tscm)

tc=ConfusionMatrixDisplay(confusion_matrix=trcm)
tc.plot()

ts=ConfusionMatrixDisplay(confusion_matrix=tscm)
ts.plot()
print("Classification report:\n",classification_report(ytest,yptest))
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import fbeta_score,confusion_matrix,ConfusionMatrixDisplay,classification_report

df=pd.read_csv('/content/drive/MyDrive/patterndata.csv')
x=df.iloc[:,2:-1]
y=df.iloc[:,1]

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)

rf=RandomForestClassifier()
rf.fit(xtrain,ytrain)
yptrain=rf.predict(xtrain)
yptest=rf.predict(xtest)

trfb=fbeta_score(ytrain,yptrain,average='weighted',beta=1)
tsfb=fbeta_score(ytest,yptest,average='weighted',beta=1)
trcm=confusion_matrix(ytrain,yptrain)
tscm=confusion_matrix(ytest,yptest)

print("Training fbeta:",trfb)
print("Testing fbeta:",tsfb)
print("Training Confusion matrix:\n",trcm)
print("Testing confusion matrix:\n",tscm)

tc=ConfusionMatrixDisplay(confusion_matrix=trcm)
tc.plot()
ts=ConfusionMatrixDisplay(confusion_matrix=tscm)
ts.plot()
print("Classification report:\n",classification_report(ytest,yptest))
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

d=pd.read_csv('/content/drive/MyDrive/iris.data')
x=d.iloc[:,:-1]
y=d.iloc[:,-1]

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=69)

knc=KNeighborsClassifier(5)
knc.fit(xtrain,ytrain)
pred=knc.predict(xtest)
cm=confusion_matrix(ytest,pred)
print(cm)
cmd=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=knc.classes_)
cmd.plot()
```
